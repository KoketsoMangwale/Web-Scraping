```python
def request_data(website):
    response = requests.get(website)
    print(response.status_code)
    return response
```

```python
def pre2022_scrape(website):
    #create a request to the site to get data
    response = request_data(website)
    
    #Soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #html result
    bid=soup.find_all('strong')
    
    #To do: 
    for t in bid:
        
        if 'Tender/Bid # :' in t.get_text():
            print(t.get_text().replace('Tender/Bid # :' , "").strip())
        elif 'Tender/Bid #:' in t.get_text():
            print(t.get_text().replace('Tender/Bid #:' , "").strip())
     
    
```
